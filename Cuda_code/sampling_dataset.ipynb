```python
# /hpc2hdd/JH_DATA/share/ysi538/PrivateShareGroup/czhangcn_mdi_dataset_2024/dataset/raw-dataset/tcga-svs-pdf
import os

def get_tcga_folders(directory):
    tcga_folders = []
    for folder_name in os.listdir(directory):
        folder_path = os.path.join(directory, folder_name)
        if os.path.isdir(folder_path) and folder_name.startswith("TCGA"):
            tcga_folders.append(folder_path)
    return tcga_folders

def get_valid_subfolders(folder_path):
    valid_subfolders = []
    for subfolder_name in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder_name)
        if os.path.isdir(subfolder_path):
            for root, _, files in os.walk(subfolder_path):
                for file in files:
                    if file.endswith(('.svs', '.tiff')):
                        file_path = os.path.join(root, file)
                        if os.path.getsize(file_path) > 50 * 1024 * 1024:  # 文件大小大于50MB
                            valid_subfolders.append(file_path)  
                            break
                if len(valid_subfolders) >= 50:
                    break
    return valid_subfolders[:50]

directory = "/hpc2hdd/JH_DATA/share/ysi538/PrivateShareGroup/czhangcn_mdi_dataset_2024/dataset/raw-dataset/tcga-svs-pdf"
tcga_folders = get_tcga_folders(directory)

result = {}
for folder in tcga_folders:
    valid_subfolders = get_valid_subfolders(folder)
    result[folder] = valid_subfolders

wsi_dirs = []

for folder, subfolders in result.items():
    print(f"Folder: {folder}")
    for subfolder in subfolders:
        print(f"  Subfolder: {subfolder}")
        wsi_dirs.append(subfolder)
```

```python
site_dict = {}
site_dict['Brain'] = ['GBM', 'LGG']
site_dict['Gastrointestinal'] = ['COAD', 'READ', 'STAD', 'ESCA']
site_dict['Gynecologic'] = ['OV', 'UCEC', 'CESC', 'UCS']
site_dict['Hematopoietic'] = ['DLBC', 'THYM']
site_dict['Melanocytic'] = ['SKCM', 'UVM']
site_dict['Urinary'] = ['KIRC', 'KIRP', 'KICH', 'BLCA']
site_dict['Prostate/Testis'] = ['PRAD', 'TGCT']
site_dict['Endocrine'] = ['ACC', 'PCPG', 'THCA']
site_dict['Liver/PB'] = ['LIHC', 'CHOL', 'PAAD']
```

```python
# Folder: /hpc2hdd/JH_DATA/share/ysi538/PrivateShareGroup/czhangcn_mdi_dataset_2024/dataset/raw-dataset/tcga-svs-pdf/TCGA-BLCA-SVS-PDF
#   Subfolder: /hpc2hdd/JH_DATA/share/ysi538/PrivateShareGroup/czhangcn_mdi_dataset_2024/dataset/raw-dataset/tcga-svs-pdf/TCGA-BLCA-SVS-PDF/TCGA-OR-A5JF-01Z-00-DX4.F77E8849-DA63-4BF9-A4A2-7A229ADFEAA7.svs

# 构建cvs文件，第一列为wsi_name，第二列为label
import os
import pandas as pd
import random

# 假设 wsi_dirs 和 result 已经定义

# 获取 wsi_names
wsi_names = []

csv_dirs = []

# 获取 labels 并展平列表
labels = []

subtypes = []

flag = 0
for folder, subfolders in result.items():
    label = folder.split('/')[-1].split('-')[1]
    for subfolder in subfolders:
        wsi_name = os.path.basename(subfolder)
        csv_dirs.append(subfolder)
        wsi_names.append(wsi_name)
        labels.append(label)
        flag = 0
        for site, subtype in site_dict.items():
            if label in subtype:
                subtypes.append(site)
                flag = 1
                break
        if flag == 0:
            subtypes.append('Others')




# 创建 DataFrame
data = {'wsi_name': wsi_names, 'label': labels, 'subtype': subtypes, 'csv_dir': csv_dirs}
df = pd.DataFrame(data)

# 保存为 CSV 文件
df.to_csv('output.csv', index=False)
```

```python
import os
import json
from tqdm import tqdm
import numpy as np

embed_cache_path = 'data/embedding_cache'
embed_cache_path = 'embedding_cache_level0'
image_names = os.listdir(embed_cache_path)
```

```python
len(image_names)
```

```python
for img_name in tqdm(image_names):
    patch_info_path_edited = os.path.join(embed_cache_path, img_name, 'patch_info_edited.json')
    if os.path.exists(patch_info_path_edited):
        continue
    patch_info_path = os.path.join(embed_cache_path, img_name, 'patch_info.json')
    with open(patch_info_path, 'r') as f:
        patch_info = json.load(f)
        # 在最前面加上图片名
        patch_info = [img_name + i for i in patch_info]
        
    # 将修改后的patch_info写回文件
    with open(patch_info_path_edited, 'w') as f:
        json.dump(patch_info, f)
```
